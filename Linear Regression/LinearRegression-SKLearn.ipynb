```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns; sns.set()

# Read in data (using Numpy)
yearsBase, meanBase = np.loadtxt('./data/5-year-mean-1951-1980.csv', delimiter=',', unpack=True)
years, mean = np.loadtxt('./data/5-year-mean-1882-2014.csv', delimiter=',', unpack=True)
```

```python
# Linear Regression (1st order polynomial) using SKlearn
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error

lin = LinearRegression() 

skyearsBase = np.array(yearsBase.reshape(-1,1)) # have to do this for 1 feature
reg = lin.fit(skyearsBase, meanBase)

r2coeffofdet = lin.score(skyearsBase, meanBase)
print('R2 score:', r2coeffofdet)   # This is the coefficient of determination.

RMSE = np.sqrt(mean_squared_error(meanBase, reg.predict(skyearsBase))) #cant get squared false to work
print('RMSE:', RMSE)

# This generates the same scatter plot as before, but adds a line plot using the function above
plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, reg.predict(skyearsBase))  #use the predict method
plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()

print('Coefficients:', reg.coef_)
print('Intercept:', reg.intercept_)
```

```python
# Linear (2nd Order Polynomial) Regression using SKLearn
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import mean_squared_error
  
poly = PolynomialFeatures(degree = 2) 
yearsBase_poly = poly.fit_transform(np.array(yearsBase.reshape(-1,1))) 

poly.fit(yearsBase_poly, meanBase)
lin2 = LinearRegression() 
reg2 = lin2.fit(yearsBase_poly, meanBase) 

r2coeffofdet = lin2.score(yearsBase_poly, meanBase)
print('R2 score:', r2coeffofdet)   # This is the coefficient of determination.

RMSE = np.sqrt(mean_squared_error(meanBase, reg2.predict(yearsBase_poly)))
print('RMSE:', RMSE)

# This generates the same scatter plot as before, but adds a line plot using the function above
plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, reg2.predict(yearsBase_poly))  #use the predict method
plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()

print('Coefficients:', reg2.coef_)
print('Intercept:', reg2.intercept_)
```

```python
# 3rd Order Polynomial Regression using SKLearn
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
  
poly = PolynomialFeatures(degree = 3) 
yearsBase_poly = poly.fit_transform(np.array(yearsBase.reshape(-1,1))) 

poly.fit(yearsBase_poly, meanBase)
lin3 = LinearRegression() 
reg3 = lin3.fit(yearsBase_poly, meanBase) 

r2coeffofdet = lin3.score(yearsBase_poly, meanBase)
print('R2 score:', r2coeffofdet)

RMSE = np.sqrt(mean_squared_error(meanBase, reg3.predict(yearsBase_poly)))
print('RMSE:', RMSE)

# This generates the same scatter plot as before, but adds a line plot using the function above
plt.scatter(yearsBase, meanBase)
plt.plot(yearsBase, reg3.predict(yearsBase_poly))  #use the predict method
plt.title('scatter plot of mean temp difference vs year')
plt.xlabel('years', fontsize=12)
plt.ylabel('mean temp difference', fontsize=12)
plt.show()

print('Coefficients:', reg3.coef_)
print('Intercept:', reg3.intercept_)
```
